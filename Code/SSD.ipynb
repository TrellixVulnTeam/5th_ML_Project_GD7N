```python
# 파일 이름 뽑아내기
import os.path

file_1 = open("../Data/EyeDetectorv2icoco/train", "w+")

for i in range(1, 11181): # 파일 수량에 맞게 변경 !!
    xml_path = 'annotations/'+'%06d' %i+'.xml'
    if os.path.isfile(xml_path):
        file_1.write('%06d' %i)
        file_1.write('\n')

file_1.close()
```

```python
import os
file_list = os.listdir('C:/jupyter_home/Machine_Learning/ML_Project_5th/5th_ML_Project/Data/EyeDetectorv2icoco/train')
file_list
```

```python
file_name = []
for file in file_list:
    if file.count(".") == 1: 
        name = file.split('.')[0]
        file_name.append(name)
    else:
        for k in range(len(file)-1,0,-1):
            if file[k]=='.':
                file_name.append(file[:k])
                break
                
file_name
```

```python
import pickle

filePath = './test.txt'
with open(filePath, 'wb') as lf:
    pickle.dump(file_name, lf)
```
